```
using Google.OrTools.LinearSolver;

Solver
    solver = new Solver("glop", Solver.OptimizationProblemType.GLOP_LINEAR_PROGRAMMING);

Variable
    x = solver.MakeVar(0, double.MaxValue, false, nameof(x)),
    y = solver.MakeVar(0, double.MaxValue, false, nameof(y));

solver.Add(3 * x + 2* y <= 6);
solver.Add(2 * x - 3 * y <= 2);

solver.Maximize(5 * x + 3 * y);

Solver.ResultStatus 
    results = solver.Solve();

display(solver.ExportModelAsLpFormat(false));
```
